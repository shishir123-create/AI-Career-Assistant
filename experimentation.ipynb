```python
# 1. Load data
import pandas as pd
jobs = pd.read_csv("jobs.csv")
skills = pd.read_csv("skills.csv")

print(jobs.head())
print(skills.head())

# 2. Simple text cleaning
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

# 3. Skill extraction test
skills_list = skills["Skill"].dropna().astype(str).str.lower().tolist()

def extract_skills(text):
    text = clean_text(text)
    return list({s for s in skills_list if s in text})

sample = "I know Python, SQL, and Machine Learning using TensorFlow"
print(extract_skills(sample))
```
